# Primeros pasos con el SDK de Claude

## Objetivos de la lección
En esta primera lección, aprenderás a:
* instalar los paquetes necesarios y autenticarte con la API
* hacer tu primera solicitud al asistente de IA Claude

## Instalando el SDK

Antes de comenzar con el SDK, asegúrate de tener Python instalado en tu sistema.  

El SDK de Claude para Python requiere Python 3.7.1 o superior.  

Puedes verificar la versión actual de Python ejecutando el siguiente comando en tu terminal:

```
python --version
```
